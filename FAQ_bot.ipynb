```
import requests 
from bs4 import BeautifulSoup 
  
URL = "http://www.values.com/inspirational-quotes"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup.prettify()) 
```

```
table = soup.find("div", { "id" : "all_quotes" })
quotes = []
for i in table.findAll(attrs= { "class" : "col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top" }):
  quote = {}
  quote["url"] = i.a["href"]
  quote["img"] = i.img["src"]
  quote["text"] = i.get_text()
  quotes.append(quote)

print(quotes)
print(quotes[0]["url"])
```

```
import csv 
filename = 'inspirational_quotes.csv'
with open(filename, 'w') as f: 
  w = csv.DictWriter(f,['url','img','text'])
  w.writeheader() 
  for quote in quotes: 
      w.writerow(quote)
```

```
with open('inspirational_quotes.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            print(f'\t{row[0]} works in the {row[1]} department, and was born in {row[2]}.')
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```
# import these modules 
from nltk.stem import WordNetLemmatizer 
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
lemmatizer = WordNetLemmatizer() 
stop_words = set(stopwords.words('english')) 
def cleaning(sentence):
  words = word_tokenize(sentence)
  words = [w for w in words if (w.isalnum() and  (w not in stop_words))]
  final_str=""
  for i in words:
    final_str+=i
    final_str+=" "
  return final_str
a = cleaning("I am very sad")
print(a)
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder as LE
from sklearn.model_selection import train_test_split 
vectorizer = CountVectorizer()
tfv = TfidfVectorizer(min_df=1, stop_words='english')
all_ques = []
for ques in test_data['Question']:
  mod_ques = cleaning(ques)
  all_ques.append(mod_ques)
tfv.fit(all_ques)
X = tfv.transform(all_ques)
"""X = vectorizer.fit_transform(all_ques)"""
print(tfv.get_feature_names())
le = LE()
Y = le.fit_transform(test_data['Class'])
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=109)
```

```
from sklearn import svm
cl = svm.SVC(kernel = 'linear')
cl.fit(X_train,y_train)
y_pred = cl.predict(X_test)
```

```
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
